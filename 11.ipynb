```python
mean = np.mean(data)
variance = np.var(data)
std_deviation = np.std(data)
```

```python
count, bins, ignored = plt.hist(data, 20, density=True, alpha=0.6, color='g', label='Эмпирическая функция распределения')
plt.plot(bins, 1/(std_deviation * np.sqrt(2 * np.pi)) * np.exp( - (bins - mean)**2 / (2 * std_deviation**2) ), linewidth=2, color='r', label='Теоретическая функция распределения')
plt.legend()
plt.show()
```

```python
plt.hist(data, bins=10, density=True, alpha=0.6, color='g', label='Гистограмма')
```

```python
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_deviation)
l='Теоретическая плотность распределения'
plt.plot(x,p, 'k', linewidth=2, label=l)
plt.legend()
plt.show()
```

```python
k_stat, k_p_value = stats.kstest(data, 'norm', args=(mean, std_deviation))
print("Критерий Колмогорова:", k_stat, "p-значение:", k_p_value)
```

```python
p_stat, p_p_value = stats.normaltest(data)
print("Критерий Пирсона:", p_stat, "p-значение:", p_p_value)
```
