```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy. stats import norm
x = np.arange (-3,5,0.2)
plt.plot(x, norm. pdf (x,0,1), label = '0,1')
plt.plot(x, norm. pdf (x,2,1), label = '2,1')
plt.plot(x, norm. pdf (x,0,0.5), label = '0,0.5')
plt.plot(x, norm. pdf (x,1,2), label = '1,2')
plt.legend()
```

```python
plt.plot(x, norm. cdf (x,0,1), label = '0,1')
plt.plot(x, norm. cdf (x,2,1), label = '2,1')
plt.plot(x, norm. cdf (x,0,0.5), label = '0,0.5')
plt.plot(x, norm. cdf (x,1,2), label = '1,2')
plt.legend()
```

```python
from scipy. stats import expon
x = np.arange (0,20,0.5)
plt.plot(x, expon. pdf (x,0,1/1), label = '1')
plt.plot(x, expon. pdf (x,0,1/0.5), label = '0.5')
plt.plot(x, expon. pdf (x,0,1/0.1), label = '0.1')
plt.legend()
```

```python
plt.plot(x, expon. cdf (x,0,1/1), label = '1')
plt.plot(x, expon. cdf (x,0,1/0.5), label = '0.5')
plt.plot(x, expon. cdf (x,0,1/0.1), label = '0.1')
plt.legend()
```

```python
from scipy. stats import binom
x = np.arange (0,25,1)
plt.bar(x, binom.pmf(x,25,0.7), label = '0.7')
plt.bar(x, binom.pmf(x,25,0.5), label = '0.5')
plt.bar(x, binom.pmf(x,25,0.2), label = '0.2')
plt.legend()
```
